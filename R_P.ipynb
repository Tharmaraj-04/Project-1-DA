```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Data Analyst internship\project\Retail Business Performance & Profitability Analysis (1).csv")
```

```python
df
```

```python
# Avoid divide-by-zero
df['Inventory_Days'] = df['Units_Sold'].apply(lambda x: 30 / x if x > 0 else np.nan)

# Correlation between Inventory Days and Profit
correlation = df[['Inventory_Days', 'Profit']].corr()
correlation_value = correlation.loc['Inventory_Days', 'Profit']
correlation_value
```

```python
# Estimate Inventory Days as a proxy: 30 / Units_Sold
df['Inventory_Days'] = df['Units_Sold'].apply(lambda x: 30 / x if x > 0 else np.nan)
```

```python
# Flag slow-moving products
slow_moving = df[df['Inventory_Days'] > 25].copy()
slow_moving['Flag'] = 'Slow-Moving'
```

```python
# Flag overstocked products with low profit
overstocked = df[(df['Inventory_Status'] == 'In Stock') & (df['Profit'] < 50)].copy()
overstocked['Flag'] = 'Overstocked'
```

```python

# Combine both flags
strategic_flags = pd.concat([slow_moving, overstocked])
```

```python
# Show key suggestions
summary = strategic_flags[[
    'Product_Category', 'Product_Name', 'Units_Sold', 'Inventory_Days',
    'Profit', 'Inventory_Status', 'Flag'
]].sort_values(by='Profit')

print(summary.head(10))
```
